```python
import pandas as pd



# Load raw data
data = pd.read_csv(r"C:\Users\kingsta\Desktop\week_two(2)\10_ACADAMY_KAIM_W2\data\Copy of Week2_challenge_data_source(CSV).csv")
print(data.head())

```

```python
for col in data.columns:
    if data[col].dtype == 'object':
        print(f"Column '{col}' contains non-numeric data.")
    else:
        try:
            pd.to_numeric(data[col])
        except ValueError:
            print(f"Column '{col}' contains mixed types.")
```

```python
data['Start'] = pd.to_datetime(data['Start'], errors='coerce')
data['End'] = pd.to_datetime(data['End'], errors='coerce')
```

```python
for col in data.select_dtypes(include='object').columns:
    data[col].fillna(data[col].mode()[0], inplace=True)
```

```python
for col in data.select_dtypes(include=['float64', 'int64']).columns:
    if data[col].isnull().sum() > 0:  # Check if there are missing values
        data[col].fillna(data[col].median(), inplace=True)  # Replace NaN with median
```

```python
# Select only numerical columns
numerical_columns = data.select_dtypes(include=['number'])

# Check skewness for numerical columns
skewness = numerical_columns.skew()
print(skewness)
```

```python
import numpy as np

# Apply log transformation for positive skewed columns
positive_skewed_columns = skewness[skewness > 1].index

for column in positive_skewed_columns:
    data[column] = np.log1p(data[column])  # log1p handles zero values as well

# Check the skewness again
new_skewness = data[positive_skewed_columns].skew()
print(new_skewness)
```

```python
# Top 10 handsets
top_10_handsets = data['Handset Type'].value_counts().head(10)
print(top_10_handsets)
```

```python
# Top 3 handset manufacturers
top_3_manufacturers = data['Handset Manufacturer'].value_counts().head(3)
print(top_3_manufacturers)
```

```python
# Filter for the top 3 manufacturers
top_3_manufacturer_names = top_3_manufacturers.index
filtered_data = data[data['Handset Manufacturer'].isin(top_3_manufacturer_names)]

# Group by Manufacturer and Handset Type
top_5_handsets_per_manufacturer = (
    filtered_data.groupby(['Handset Manufacturer', 'Handset Type'])
    .size()
    .reset_index(name='count')
    .sort_values(['Handset Manufacturer', 'count'], ascending=[True, False])
)

# Get the top 5 handsets per manufacturer
top_5_handsets_per_manufacturer = (
    top_5_handsets_per_manufacturer.groupby('Handset Manufacturer')
    .head(5)
)
print(top_5_handsets_per_manufacturer)
```
